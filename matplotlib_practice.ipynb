```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis,y_axis)
```

```python
plt.plot(y_axis,x_axis, 'o', c ='red')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
#Now Making a bubble chart for the first time
plt.scatter(x_axis,y_axis)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Now making the y axis data larger
y_axis_larger =[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis,y_axis, s=y_axis_larger)
```

```python
# Format for list comprehension
#new_list = [expression for item in list if conditional]
```

```python
# Now using l comp in the scatterplot (bubble) function
plt.scatter(x_axis,y_axis, s= [i*3 for i in y_axis])
```

```python
#Now do this shit AGAIN but using the stupid object oriented way.
fig,ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
# Now for a bubble chart
fig,ax = plt.subplots()
ax.scatter(x_axis,y_axis, s=y_axis)
```

```python
#SKILL CHALLENGE: Do a bunch of stuff, but also use the object oriented programming method. Also change edgecolors, make the colors 20% more transparent, invert y axis, etc. 

fig,ax =plt.subplots()
ax.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], color='skyblue', label ='Boston', edgecolors= 'black', alpha = 0.80)
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
ax.invert_yaxis()
ax.set_xlim(0,50)
ax.legend()
ax.set_title("PyBer Fare by Month")
```

```python
# using MATLAB for a pie chart
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors: One for each month
colors = ["slateblue","magenta","lightblue","green","yellowgreen",'greenyellow',"yellow","orange", "gold", 'indianred',"tomato","mistyrose"]
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,explode=explode_values,colors=colors,labels=x_axis,autopct = '%.1f%%')

plt.show()
```
